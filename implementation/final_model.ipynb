```python
X = X.reshape(X.shape[0], 1, X.shape[1])
```

```python
X.shape[2]
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
```

```python
# Define the model
model = Sequential([
    LSTM(64, activation='tanh', return_sequences=False, input_shape=(1, X.shape[2])),
    Dropout(0.2),
    Dense(32, activation='relu'),
    Dense(3, activation='softmax')  # Still use softmax for multi-class classification
])
```

```python
# Compile the model (use sparse_categorical_crossentropy)
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Model summary
model.summary()
```

```python
history = model.fit(X, y, epochs=10, batch_size=32, validation_split=0.2, verbose=1)
```

```python

from tensorflow.keras.utils import plot_model

plot_model(model, to_file='model_plot.png', show_shapes=True, show_layer_names=True)
```

```python
import numpy as np
```

```python
new_pred = model.predict(X[:5])
predicted_labels = np.argmax(new_pred, axis=1)  # Convert probabilities to class labels
print("Predicted labels:", predicted_labels)
print("actual: ", y[:5])
```

```python
test_loss, test_acc = model.evaluate(X[:10000], y[:10000], verbose=1)
print(f"Test Accuracy: {test_acc:.4f}")
print(f"Test Loss: {test_loss:.4f}")
```

```python
from sklearn.metrics import confusion_matrix, classification_report

y_pred = np.argmax(model.predict(X[:10000]), axis=1)
print(confusion_matrix(y[:10000], y_pred))
print(classification_report(y[:10000], y_pred))
```
